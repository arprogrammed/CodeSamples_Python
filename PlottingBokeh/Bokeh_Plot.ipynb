```python
import bokeh
import pandas
from bokeh.plotting import figure
from bokeh.io import output_file, show

df = pandas.read_csv('bachelors.csv')

x = df['Year']
y = df['Engineering']

output_file('graphline.html')

f = figure(plot_width=500, plot_height=400)
f.title.text="Cool Data"
f.title.text_color="Gray"
f.title.text_font="times"
f.title.text_font_style="bold"

f.xaxis.minor_tick_line_color=None
f.yaxis.minor_tick_line_color=None
f.xaxis.axis_label="Date"
f.yaxis.axis_label="Intensity"    

f.line(x,y, line_color='blue')

show(f)
```

```python
import bokeh
import pandas
from bokeh.plotting import figure
from bokeh.io import output_file, show
from bokeh.models import ColumnDataSource

df = pandas.read_csv('adbe.csv', parse_dates=['Date'])
df['Volume'] = df['Volume'] * 0.000001

source = ColumnDataSource(data = df)

f = figure(x_axis_type='datetime', sizing_mode = 'stretch_both')

f.vline_stack(['Close', 'Volume'], x='Date', source=source, line_color=['green','orange'], line_width=2)

f.title.text="Adobe Stock Price"
f.title.text_color="Gray"
f.title.text_font="times"
f.title.text_font_style="bold"


output_file('timeseries.html')

show(f)
```

```python
import pandas
import bokeh
from bokeh.plotting import figure
from bokeh.io import output_file, show
from bokeh.models import HoverTool, ColumnDataSource

df = pandas.read_csv('SPT.csv', parse_dates=['Start', 'End'])

df['Start_string']=df['Start'].dt.strftime('%Y-%m-%d %H:%M:%S')
df['End_string']=df['End'].dt.strftime('%Y-%m-%d %H:%M:%S')

source = ColumnDataSource(data = df)

p = figure(x_axis_type='datetime', sizing_mode = 'stretch_both')

q = p.quad(left='Start', right='End', bottom=0, top=1, fill_color='green', source=source)

p.yaxis.minor_tick_line_color=None
p.yaxis.ticker.desired_num_ticks=1

hover = HoverTool(tooltips = [('Start', '@Start_string'),('End', '@End_string')])
p.add_tools(hover)

p.title.text="Motion Graph"
p.title.text_color="Gray"
p.title.text_font="times"
p.title.text_font_style="bold"

output_file('timeplot.html')

show(p)
```
